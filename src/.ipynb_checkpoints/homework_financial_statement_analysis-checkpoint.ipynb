## 5 Let's assume that the target company will set the dividend payout ratio to 100% 
## from next year onward, and the latest profit structure will not change forever in the future.  
## Assuming that the cost of capital of the target company's stock is permanently
## 　constant at the return on equity in the most recent fiscal year, what is its stock value?

```python
import pandas as pd
#The target company　 this time is ZHD(4689).
financial_df = pd.read_csv("financial_4689.csv", encoding="shift-jis")
financial_df 
```

```python
dividend_per_share= float(financial_df["一株当たり配当金"][0:1].values[0])
print(dividend_per_share)
total_number_of_issued_shares = financial_df["発行済株式総数"][0:1].values[0]
print(total_number_of_issued_shares)
#The unit is 1 million yen
dividend = dividend_per_share * total_number_of_issued_shares /(10000 *100)
print(dividend)
roe = financial_df["ROE"][0:1].values[0] 
print(roe)
print(round(dividend/roe, 2))
```

## 7.If the cost of capital of the target company 
## 　is the return on equity in the most recent fiscal year, 
## 　　what is the weighted average cost of capital if the tax rate is 30%?

```python
# ①負債合計・・・2886086(百万円)
sum_debt = 2886086
# ②純資産合計・・・1047823(百万円)
sum_net_worth = 2886086
# ③負債コスト・・・0.16%
debt_cost = 0.16/100
# ④株式資本コスト・・・ 4.6%
stock_cost = 4.6/100
# ⑤法人税率・・・30%
tax_rate = 30/100
# wacc = 負債合計/(負債合計+純資産合計) * 負債コスト * (1 - 法人税率) 
#           + 純資産合計/(負債合計 + 純資産合計) * 株式資本コスト

wacc = (sum_debt/(sum_debt + sum_net_worth) * debt_cost * (1 - tax_rate) + sum_net_worth /(sum_debt + sum_net_worth ) * stock_cost )*100
wacc
```
